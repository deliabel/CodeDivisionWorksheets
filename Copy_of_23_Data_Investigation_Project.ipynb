```python
import pandas as pd
url = 'https://raw.githubusercontent.com/futureCodersSE/working-with-data/main/Data%20sets/housing_in_london_yearly_variables.csv'
housingdf = pd.read_csv(url)
housingdf
```

```python
#housingdf['area']
area_names = housingdf['area'].unique()
for each_area in area_names:
  print(each_area)
print(len(area_names), 'areas')
```

```python
dates_collected = housingdf['date'].unique()
for year in dates_collected:
  print(year)
print(len(dates_collected), 'years')
```

```python
print('51 areas over 21 years, 51 x 21 =', 51*21, '\n')

print(housingdf.info())
```

```python
# area_names = housingdf['area'].unique() not a list... numpy array? check video again

# display all records about one area: print certain rows
# display not ideal, each as a row like above would be better

areas_list = []
for each_area in area_names:
  areas_list.append(each_area)
print(areas_list)

def get_areaInfo(area_name):
  area_index = areas_list.index(area_name)
  look_up = 51 - area_index
  for n in range(1, 22):
    print(housingdf.iloc[(51*n) - look_up])

get_areaInfo('merton')
```
